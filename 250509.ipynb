```python
month= ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
rain= [30, 40, 60, 80, 100, 150, 300, 280, 120, 70, 50, 35]
plt.figure(figsize=(12,6))
plt.title('monthly rain')
plt.xlabel('month')
plt.ylabel('rain(mm)')
plt.plot(month, rain,color='c',linestyle=':',marker='o')
for i in range(len(month)):
  plt.annotate(rain[i],xy=(month[i],rain[i]),xytext=(5,-10),textcoords='offset points')
plt.show()
```

```python
month= ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
rain= [30, 40, 60, 80, 100, 150, 300, 280, 120, 70, 50, 35]
plt.figure(figsize=(12,6))
plt.title('monthly rain')
plt.xlabel('month')
plt.ylabel('rain(mm)')
plt.plot(month, rain,color='c',linestyle=':',marker='o')

#데이터포인트에 값 표시
for i,value in enumerate(rain):
  plt.text(i, value+10, str(value), ha='center',va='bottom')

#레이아웃 조정
plt.tight_layout()

plt.show()
```
