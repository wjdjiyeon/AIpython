```python
q1_sw=vir_data['SepalWidthCm'].quantile(0.25)
q3_sw=vir_data['SepalWidthCm'].quantile(0.75)
iqr_sw=q3_sw-q1_sw
print(q1_sw,q3_sw.round(2),iqr_sw.round(2))
```

```python
lower_bound_sw=q1_sw-1.5*iqr_sw
upper_bound_sw=q3_sw+1.5*iqr_sw
print(lower_bound_sw.round(2),upper_bound_sw.round(2))
```

```python
sw_out=vir_data[(vir_data['SepalWidthCm']<lower_bound_sw) | (vir_data['SepalWidthCm']>upper_bound_sw)]
print("꽃받침 너비 이상치:")
sw_out
```

```python
no_out1=vir_data[(vir_data['SepalWidthCm']>=lower_bound_sw) & (vir_data['SepalWidthCm']<=upper_bound_sw)]
print(len(no_out1))
```
