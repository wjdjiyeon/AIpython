```python
df.describe()
```

```python
q1_score=df['점수'].quantile(0.25)
q3_score=df['점수'].quantile(0.75)
iqr_score=q3_score-q1_score
print("q1_score: ",q1_score,"q3_score: ",q3_score,"iqr_score: ",iqr_score)
```

```python
lower_bount_score=q1_score-1.5*iqr_score
upper_bount_score=q3_score+1.5*iqr_score
print("하한경계,상한경계: ",lower_bount_score,upper_bount_score)
```

```python
score_outliers=df[(df['점수']<lower_bount_score) | (df['점수']>upper_bount_score)]
print("점수 이상치:")
print(score_outliers)
```

```python
q1_study=df['공부시간'].quantile(0.25)
q3_study=df['공부시간'].quantile(0.75)
iqr_study=q3_study-q1_study
print("q1_study: ",q1_study,"q3_study: ",q3_study,"iqr_study: ",iqr_study)
```

```python
study_outliers=df[(df['공부시간']<q1_study-1.5*iqr_study) | (df['공부시간']>q3_study+1.5*iqr_study)]
print("공부시간 이상치:")
print(study_outliers)
```

```python
df_no_outliers=df[
    (df['점수']>=lower_bount_score) & (df['점수']<=upper_bount_score) &
    (df['공부시간']>=q1_study-1.5*iqr_study) & (df['공부시간']<=q3_study+1.5*iqr_study)
]
```

```python
df_no_outliers=df_no_outliers.reset_index()
print("이상치가 제거된 데이터프레임")
print(df_no_outliers)
```

```python
#문제 11: 데이터 그룹핑
class_mean=df_no_outliers.groupby('분반')['점수'].mean()
print("반별 점수 평균:")
print(class_mean.round(2))
```

```python
#문제 12: 데이터 시각화
plt.figure(figsize=(10,6))
plt.bar(class_mean.index,class_mean.values,color='skyblue',width=0.6)
plt.title('class eman', fontsize=16)
plt.xlabel('class', fontsize=14)
plt.ylabel('mean score', fontsize=14)
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
#공부시간과 점수의 산점도 그래프
plt.figure(figsize=(10,6))
plt.scatter(df_no_outliers['공부시간'],df_no_outliers['점수'],color='skyblue')
plt.title('study time vs score', fontsize=16)
plt.xlabel('study time', fontsize=14)
plt.ylabel('score', fontsize=14)
plt.grid(alpha=0.3)
plt.show()
```

```python
data = pd.Series([10, 11, 12, 13, 14, 15, 100, 101])
```

```python
data
```

```python
q1 = data.quantile(0.25)
q3=data.quantile(0.75)
iqr = q3 - q1
print("q1: ",q1,"q3: ",q3,"iqr: ",iqr)
```

```python
lower_bound= q1 - 1.5*iqr
upper_bound= q3 + 1.5*iqr
print("하한경계,상한경계: ",lower_bound,upper_bound)
```

```python
#2 이상치 판별 기준

#하한 경계 미만의 값: 이상치
#상한 경계 초과의 값: 이상치

#3.3 이상치 탐지 과정

#데이터에서 Q1, Q3, IQR 계산
#하한 경계와 상한 경계 계산
#경계를 벗어나는 값 식별하고 제거
```

```python
for i in data:
    if i < lower_bound or i > upper_bound:
        print(i)
```
