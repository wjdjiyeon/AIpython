```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
temperatures = [5, 7, 12, 18, 22]

plt.plot(months, temperatures) # x축 데이터, y축 데이터 순서로 입력
plt.title('Monthly Temperature Change') # 그래프 제목
plt.xlabel('Month')                 # x축 이름
plt.ylabel('Temperature (C)')       # y축 이름
plt.show()                          # 그래프 보여주기
```

```python
year = [2006, 2009, 2012, 2015, 2018]
kor = [547, 546, 554, 524, 526]
usa = [474, 487, 481, 470, 478]
plt.plot(year,kor,label='kor')
plt.plot(year,usa,label='usa')
plt.xlabel('year')
plt.ylabel('country')
plt.legend()
plt.show()
```

```python
year = [2006, 2009, 2012, 2015, 2018]
kor = [547, 546, 554, 524, 526]
usa = [474, 487, 481, 470, 478]
jpn = [523, 529, 536, 532, 527]

# 그래프가 가로와 세로 크기를 각각 5인치와 3인치 크기 설정
plt.figure(figsize=(5,3))
# year 데이터를 x축으로, kor 데이터를 y축으로 하여 빨간색 점선으로 선 그래프를 그리고 범례에 'Korea'라고 표시
plt.plot(year, kor,'r:', label='korea' )
plt.plot(year, usa, 'b:',label='usa' )
plt.plot(year, jpn, 'g-', label='Japan')
plt.ylim(400, 600)  # y축 범위 설정
# 그래프에 범례를 추가
plt.legend()
plt.title('Mathematics performance') # 그래프 제목을 'Mathematics performance'로 설정합니다.
plt.xlabel('year')
plt.ylabel('score')
plt.show()
```

```python
month= ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
rain= [30, 40, 60, 80, 100, 150, 300, 280, 120, 70, 50, 35]
plt.figure(figsize=(12,6))
plt.title('monthly rain')
plt.xlabel('month')
plt.ylabel('rain(mm)')
plt.plot(month, rain,color='c',linestyle=':',marker='o')
for i in range(len(month)):
  plt.annotate(rain[i],xy=(month[i],rain[i]),xytext=(5,-10),textcoords='offset points')
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.title('monthly rain')
plt.xlabel('month')
plt.ylabel('rain(mm)')
plt.plot(month, rain,'g.-') #색,마커,라인스타일
plt.show()
```
